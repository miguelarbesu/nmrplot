# Examples using `nmrplot` in a notebook

```python
import nmrplot.core as npl

%load_ext autoreload
%autoreload 2
```

```python
experiment_path = "../data/bruker_2d_heteronuclear"
spectrum = npl.Spectrum(experiment_path)
spectrum.threshold = 10
spectrum.get_ppm_ranges()
```

```python
fig, ax = spectrum.plot_spectrum(linewidth=0.25, xlims=(12, 0), ylims=(200, 0))
```

```python
fig.savefig("./test.png", dpi=300)
```
